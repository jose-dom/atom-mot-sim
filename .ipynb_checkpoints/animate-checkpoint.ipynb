```python
# importing necessary packages
import itertools
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import display, clear_output
%matplotlib inline
```

```python
# 2D animation
# defining the x and y boundaries
xmin, xmax = 0, 4*np.pi
ymin, ymax = -1, 1

# creating axes of time steps needed for the function
t = np.linspace(xmin,xmax)
# function operating through time
x = np.sin(t)

# defining the figure and axes
fig, ax = plt.subplots()
# creating the line within the x and y boundaries
l, = ax.plot([xmin,xmax],[ymin,ymax])

# utilizing a lambda function to add data for each time step in the function
animate = lambda i: l.set_data(t[:i], x[:i])

# animating the function
for i in range(len(x)):
    animate(i)
    clear_output(wait=True)
    display(fig)

# displaying the final figure
plt.show()
```

```python
# 3D animation
# defining the x and y boundaries
xmin, xmax = 0, 4*np.pi
ymin, ymax = -1, 1
zmin, zmax = -1, 1

# creating axes of time steps needed for the function
t = np.linspace(xmin,xmax)
# function operating through time
x = np.sin(t)
y = np.cos(t)

# defining the figure and axes
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# creating the line within the x and y boundaries
l, = ax.plot([xmin,xmax],[ymin,ymax],[zmin,zmax])

# utilizing a lambda function to add data for each time step in the function
animate = lambda i: l.set_data(t[:i], x[:i])
animateZ = lambda j: l.set_3d_properties(y[:i])

# animating the function
for i in range(len(x)):
    animate(i)
    animateZ(i)
    clear_output(wait=True)
    display(fig)

# displaying the final figure
plt.show()
```
